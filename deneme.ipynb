```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def olay_gerceklesti(event , x,y ,flags,param):
    if event == cv2.EVENT_LBUTTONDOWN:
        cv2.circle(bos_resim,(x,y), 50, (127,127,127),-1)
    elif event == cv2.EVENT_RBUTTONDOWN:
        cv2.circle(bos_resim,(x,y), 50, (50,50,50),-1)
    pass

cv2.namedWindow("ResimPenceresi")
cv2.setMouseCallback("ResimPenceresi", olay_gerceklesti)
bos_resim = np.zeros((1000,1000,3), dtype=np.uint8)

while True:
   cv2.imshow("ResimPenceresi", bos_resim)
   key = cv2.waitKey(10)
   if key == ord("q"):
       cv2.destroyAllWindows()
       break


# cv2.imshow("ResimPenceeresi", bos_resim)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```
