```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def olay_gerceklesti(event , x,y ,flags,param):
    if event == cv2.EVENT_LBUTTONDOWN:
        cv2.circle(bos_resim,(x,y), 50, (127,127,127),-1)
    elif event == cv2.EVENT_RBUTTONDOWN:
        cv2.circle(bos_resim,(x,y), 50, (50,50,50),-1)
    pass

cv2.namedWindow("ResimPenceresi")
cv2.setMouseCallback("ResimPenceresi", olay_gerceklesti)
bos_resim = np.zeros((1000,1000,3), dtype=np.uint8)

while True:
   cv2.imshow("ResimPenceresi", bos_resim)
   key = cv2.waitKey(10)
   if key == ord("q"):
       cv2.destroyAllWindows()
       break


# cv2.imshow("ResimPenceeresi", bos_resim)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
resim = cv2.imread("emre.jpg")

matris = np.array([[0, -1, 0],
                   [-1, 5, -1],
                   [0, -1, 0]])
resim = cv2.filter2D(resim,-1,matris)

cv2.imwrite("emre.jpg",resim)
```

```python
ilresim = cv2.imread("deneme.jpg")
kopyasi = ilresim.copy()
matris = np.array([[0, -1, 0],
                   [-1, 5, -1],
                   [0, -1, 0]])

for satir in range(1, ilresim.shape[0]-1):
    for sutun in range(1, ilresim.shape[1]-1):
        deger = np.sum(np.multiply(kopyasi[satir-1:satir+2, sutun-1:sutun+2], matris))
        kopyasi[satir, sutun] = deger

cv2.imshow("ilhali", ilresim)
cv2.imshow("dnsms", kopyasi)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
